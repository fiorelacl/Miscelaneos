```python
def class_events_NDJ34(ds,name):
    ## It's manly used with CESM2_Historical y GFDL_Historical (a.k.a FLOR_Historical)
    
    gb = ds.sst.groupby('time.month')
    ds_a = gb - gb.mean(dim='time')
    index3 = ds_a.mean(dim=['lon', 'lat'])
    index34=index3.rolling(time=3, center=True).mean()
    index34_NDJ=index34.sel(time=index34.time.dt.month.isin([12])).reset_coords('month', drop=True)
    # std_dev = index34_NDJ.std()
    # norm_index34_NDJ = np.round(index34_NDJ/std_dev,2)

    df=index34_NDJ.to_dataframe().reset_index()

    df['Source'] = name
    df['Year'] = df['time'].dt.year
    df['Preceeding'] = df['sst'].shift(1)
    df['W1'] = df['sst'].shift(0)
    df['W2'] = df['sst'].shift(-1)
    df['W3'] = df['sst'].shift(-2)
    df['W4'] = df['sst'].shift(-3)
    df['Strong_Nino'] = df['Preceeding'].apply(lambda x: 1 if x > 1.5 else 0)
    
    conditions = [
        (df['W1'] < -0.5) & (df['W2'] < -0.5) & (df['W3'] < -0.5)  & (df['W4'] < -0.5),  # 4th 
        (df['W1'] < -0.5) & (df['W2'] < -0.5) & (df['W3'] < -0.5),  # Triple-dip
        (df['W1'] < -0.5) & (df['W2'] < -0.5),  # Double-dip
        (df['W1'] < -0.5),  # Single La NiÃ±a
        (df['W1'] >= -0.5)
    ]
    classifications = ['4More','TLN', 'DLN', 'SLN','None']
    
    df['Classification'] = np.select(conditions, classifications, default='None')

    df['Classification'] = df['Classification'].where(~df['Classification'].shift(1).isin(['4More','TLN', 'DLN']), 'None')
    df['Classification'] = df['Classification'].where(~df['Classification'].shift(2).isin(['TLN','DLN']), 'None')
    df['Classification'] = df['Classification'].where(~df['Classification'].shift(3).isin(['TLN']), 'None')

    df['New_Class'] = np.where(df['Classification'].isin(['4More','TLN', 'DLN']), np.where(df['Strong_Nino'] == 1, 'MLN_wSEN', 'MLN_woSEN'), df['Classification'])
    
    dff=df.drop(['time', 'sst','Strong_Nino'], axis=1)
    dff= df[['Year', 'Source',	'Preceeding',	'W1',	'W2',	'W3',	'W4',	'Classification',	'New_Class']]

    return dff
```

```python
def class_events_NDJ(ds,name,y0):
    # It is mainly used with experiments, but we apply it to HadISST first due to the fileâs structure. 
    # So, for that specific file, we just comment or uncomment the indicated lines as needed.
    
    std_dev = ds.std() # Comment this when running HaddIST
    norm_index34_NDJ = np.round(ds/std_dev,2) # Comment this when running HaddIST
    #norm_index34_NDJ = ds # Unccomment this when running HaddIST
    df=norm_index34_NDJ.to_dataframe().reset_index().set_axis(['Year','sst'], axis=1)

    df['Year'] = df['Year'] + y0
    df['Source'] = name
    df['Preceeding'] = df['sst'].shift(1)
    df['W1'] = df['sst'].shift(0)
    df['W2'] = df['sst'].shift(-1)
    df['W3'] = df['sst'].shift(-2)
    df['W4'] = df['sst'].shift(-3)
    df['Strong_Nino'] = df['Preceeding'].apply(lambda x: 1 if x > 1.5 else 0)
    
    conditions = [
        (df['W1'] < -0.5) & (df['W2'] < -0.5) & (df['W3'] < -0.5)  & (df['W4'] < -0.5),  # 4th or more
        (df['W1'] < -0.5) & (df['W2'] < -0.5) & (df['W3'] < -0.5),  # Triple-dip
        (df['W1'] < -0.5) & (df['W2'] < -0.5),  # Double-dip
        (df['W1'] < -0.5),  # Single La NiÃ±a
        (df['W1'] >= -0.5) #None classification
    ]
    classifications = ['4More','TLN', 'DLN', 'SLN','None']
    
    df['Classification'] = np.select(conditions, classifications, default='None')

    df['Classification'] = df['Classification'].where(~df['Classification'].shift(1).isin(['4More','TLN', 'DLN']), 'None')
    df['Classification'] = df['Classification'].where(~df['Classification'].shift(2).isin(['TLN','DLN']), 'None')
    df['Classification'] = df['Classification'].where(~df['Classification'].shift(3).isin(['TLN']), 'None')

    df['New_Class'] = np.where(df['Classification'].isin(['4More','TLN', 'DLN']), np.where(df['Strong_Nino'] == 1, 'MLN_wSEN', 'MLN_woSEN'), df['Classification'])
    
    dff=df.drop(['sst','Strong_Nino'], axis=1)

    return dff
```
